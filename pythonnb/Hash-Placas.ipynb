```python
original_df['placa-hash'] = [hashlib.sha1(str.encode(str(i))).hexdigest() for i in original_df['Placa']]

### Hash the 'placa' column
```

```python
columns = list(original_df.columns)
columns[columns.index('Placa')] = columns[columns.index('placa-hash')]
del columns[-1]

### Use this to remove the original 'placa' column
```

```python
hash_df = original_df[columns]
```
