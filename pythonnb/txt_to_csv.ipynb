```python
# Hash das placas
hash_data = new_data
```

```python
hash_data['placa-hash'] = [hashlib.sha1(str.encode(str(i))).hexdigest() for i in new_data['placa']]
```

```python
columns = list(new_data.columns)
columns[columns.index('placa')] = columns[columns.index('placa-hash')]
del columns[-1]
```
